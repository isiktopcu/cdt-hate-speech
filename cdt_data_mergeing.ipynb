```python
import pandas as pd
import numpy as np
import os
import json
import re
```

```python
# Concatenating the Text Data
dir_path = '/content/drive/MyDrive/cdt-data-latest/weeklyrevised'
all_data = []

for filename in os.listdir(dir_path):
    if filename.endswith('.json'):
        file_path = os.path.join(dir_path, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            all_data.extend(data)
output_file = '/content/drive/MyDrive/cdt-data-latest/weeklyrevised/concatenated_text_data.json'

with open(output_file, 'w') as outfile:
    json.dump(all_data, outfile)

print(f"Concatenated JSON saved to {output_file}")
```

```python
analysis['ids_isik'] = analysis['displayUrl'].apply(lambda x: re.search(r'tweetdisplay/(\d+)', str(x)).group(1) if pd.notnull(x) else None)
analysis['ids_isik'] = analysis['ids_isik'].astype(float).astype('Int64')
print(analysis[['displayUrl', 'ids_isik']])
```

```python
analysis.rename(columns={"ids_isik":"id"},inplace=True)
```

```python
analysis.columns
```

```python
duplicates_text = text[text.duplicated(['id'], keep=False)]
duplicates_analysis = analysis[analysis.duplicated(['id'], keep=False)]

print("Duplicates in 'text' DataFrame:")
print(duplicates_text)

print("Duplicates in 'analysis' DataFrame:")
print(duplicates_analysis)

text_unique = text.drop_duplicates(subset='id', keep='first')
analysis_unique = analysis.drop_duplicates(subset='id', keep='first')

merged_df = pd.merge(text_unique, analysis_unique, on='id')
```
