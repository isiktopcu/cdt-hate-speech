```python
import pandas as pd
import numpy as np
```

```python
busra = pd.read_excel("/content/drive/MyDrive/busra/Pittsburgh_CDT_hate_speech_annotation_240717_sample_tweets_annotation_busra.xlsx")
isik = pd.read_excel("/content/drive/MyDrive/busra/Pittsburgh_CDT_hate_speech_annotation_240717_sample_tweets_annotation_isik_busra_isik_annotated.xlsx")
```

```python
busra = busra.astype(str)
```

```python
busra.columns
```

```python
columns_to_convert = ['O_Neutral','Annotation #']
for col in columns_to_convert:
    busra[col] = pd.to_numeric(busra[col], errors='coerce').astype('Int64')  # Using 'Int64' type to keep NaN as NaN
```

```python
busra['O_Neutral'] = busra['O_Neutral'].fillna(1)
```

```python
columns_to_convert = ['O_relevant','O_Physical_Harm_Threat',
       'O_Insult_profanity', 'O_Demeaning', 'O_Insult_(Ridicule Mockery)',
       'O_Other', 'Hate Speech', 'H_Gender', 'H_Sexual_orientation',
       'H_Religion_Belief', 'H_Ideology', 'H_Ethnicity', 'H_Other']
for col in columns_to_convert:
    isik[col] = pd.to_numeric(isik[col], errors='coerce').astype('Int64')
```

```python
isik
```

```python
isik['O_relevant'] = isik['O_relevant'].fillna(0)
```

```python
columns = ['O_Physical_Harm_Threat',
       'O_Insult_profanity', 'O_Demeaning', 'O_Insult_(Ridicule Mockery)',
       'O_Other', 'Hate Speech', 'H_Gender', 'H_Sexual_orientation',
       'H_Religion_Belief', 'H_Ideology', 'H_Ethnicity', 'H_Other']
for col in columns:
  isik[col] = isik[col].fillna(0)
```

```python
isik["annotator"] = "isik"
```

```python
propagate_columns = ['Annotation #', '_id', 'Twitter/X Account', 'Candidate Name',
                     'Candidate Party Affiliation', 'Incumbant Status', 'text']


adjudicator_dummy = pd.DataFrame(np.nan, index=range(len(busra)), columns=busra.columns)
adjudicator_dummy['annotator'] = 'adjudicator'
for col in propagate_columns:
    adjudicator_dummy[col] = busra[col].values  # Ensure correct data propagation

# Concatenate the dataframes in the specified order: busra, isik, adjudicator
df_concatenated = pd.concat([busra, isik, adjudicator_dummy]).sort_index(kind='merge')

df_final = pd.DataFrame()
for idx in range(0, len(df_concatenated), 3):
    df_final = pd.concat([df_final, df_concatenated.iloc[idx:idx+3]], ignore_index=True)

annotation_columns = [col for col in df_final.columns if col.startswith(('O_', 'H_'))]

for col in annotation_columns:
    df_final.insert(df_final.columns.get_loc(col) + 1, 'disagreement_' + col, np.nan)

df_final[annotation_columns] = df_final[annotation_columns].apply(pd.to_numeric, errors='coerce', downcast='integer')

for col in annotation_columns:
    for idx in range(2, len(df_final), 3):
        busra_value = df_final.at[idx - 2, col]
        isik_value = df_final.at[idx - 1, col]

        if pd.isna(busra_value) or pd.isna(isik_value):
            disagreement = np.nan
        elif busra_value == isik_value:
            disagreement = 0 if busra_value == 0 else 2
        else:
            disagreement = 1

        df_final.at[idx, 'disagreement_' + col] = disagreement
```
